```python
import pandas as pd
```

```python
train = pd.read_csv("../data/raw_data/train.csv")
test = pd.read_csv("../data/raw_data/test.csv")
```

```python
train = train.drop(columns = ['ID'])
test = test.drop(columns = ['ID'])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train['SUBCLASS'] = le.fit_transform(train['SUBCLASS'])
```

```python
train[test.columns] = train[test.columns].map(lambda x: 0 if x == 'WT' else 1)
test[test.columns] = test[test.columns].map(lambda x: 0 if x == 'WT' else 1)
```

# Variance Threshold
- 분산 임계값을 충족하지 않는 모든 피처를 제거하는 방식

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
VT = VarianceThreshold(threshold=(0.02)) # 0.06, 0.05, 0.04, 0.03, 0.02
```

```python
print(VT.fit_transform(train[test.columns]))
```

```python
VT.get_feature_names_out()
```

```python
VT.transform(test)
```

```python
vt_col = VT.get_feature_names_out().tolist()
```

```python
vt_col.append("SUBCLASS")
```

```python
train_vt = train[vt_col]
test_vt = test[VT.get_feature_names_out()]
```

```python
train_vt.head()
```

```python
test_vt.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(train[VT.get_feature_names_out()], train_vt['SUBCLASS'], test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# 테스트 데이터로 예측
y_pred = model.predict(x_test)
```

```python
# 평가
# 1. 정확도 (accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
test_predict = model.predict(test_vt.values)
```

```python
test_predict = le.inverse_transform(test_predict)
```

```python
submission = pd.read_csv("../data/raw_data/sample_submission.csv")
```

```python
submission['SUBCLASS'] = test_predict
```

```python
submission.to_csv("../data/submission/submission_06.csv", index=False)
```
